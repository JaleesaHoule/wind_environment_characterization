```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import astropy
from astropy import stats
import figurefirst as fifi
import pynumdiff
import math
```

```python
df=pd.read_hdf('noisetest.hdf')
# data that was collected in wind tunnel at 3 speeds. 
# ignore first several minutes of data ~[0:4000], during 
# which time the wind tunnel was not yet closed.
```

```python
def unwrap_angle (directionarray, degrees=True):
    if (degrees==False):
        directionarray=+np.pi
    else:
        newdirection=directionarray*np.pi/180-np.pi
    return np.unwrap(newdirection)

def rewrap_angle(unwrappedarray, degrees=False):
    rewrapped = (unwrappedarray + np.pi) % (2 * np.pi)
    #same as np.arctan(np.cos(unwrappedarray), np.sin(unwrappedarray))
    if degrees:
        return rewrapped*180/np.pi 
    else:
        return rewrapped #this is in radians not degrees
   
```

```python
#butter  - light: [2,5] mod: [5,.1]

nowind_speed_light, throwaway = pynumdiff.smooth_finite_difference._smooth_finite_difference.butterdiff(x=df.S2.iloc[4000:10000], dt=.1, params=[3,.5,10], options={'iterate': True})
nowind_speed_mod, throwaway = pynumdiff.smooth_finite_difference._smooth_finite_difference.butterdiff(x=df.S2.iloc[4000:10000], dt=.1, params=[3,.1,10], options={'iterate': True})

lowwind_speed_light, throwaway = pynumdiff.smooth_finite_difference._smooth_finite_difference.butterdiff(x=df.S2.iloc[12000:18000], dt=.1, params=[3,.5,10], options={'iterate': True})
lowwind_speed_mod, throwaway = pynumdiff.smooth_finite_difference._smooth_finite_difference.butterdiff(x=df.S2.iloc[12000:18000], dt=.1, params=[3,.1,10], options={'iterate': True})

modwind_speed_light, throwaway = pynumdiff.smooth_finite_difference._smooth_finite_difference.butterdiff(x=df.S2.iloc[21000:27000], dt=.1, params=[3,.5,10], options={'iterate': True})
modwind_speed_mod, throwaway = pynumdiff.smooth_finite_difference._smooth_finite_difference.butterdiff(x=df.S2.iloc[21000:27000], dt=.1, params=[3,.1,10], options={'iterate': True})



nowind_dir_light, throwaway = pynumdiff.smooth_finite_difference._smooth_finite_difference.butterdiff(x=unwrap_angle(df.D.iloc[4000:10000]), dt=.1, params=[3,.5,10], options={'iterate': True})
nowind_dir_mod, throwaway = pynumdiff.smooth_finite_difference._smooth_finite_difference.butterdiff(x=unwrap_angle(df.D.iloc[4000:10000]), dt=.1, params=[3,.1,10], options={'iterate': True})

lowwind_dir_light, throwaway = pynumdiff.smooth_finite_difference._smooth_finite_difference.butterdiff(x=unwrap_angle(df.D.iloc[12000:18000]), dt=.1, params=[3,.5,10], options={'iterate': True})
lowwind_dir_mod, throwaway = pynumdiff.smooth_finite_difference._smooth_finite_difference.butterdiff(x=unwrap_angle(df.D.iloc[12000:18000]), dt=.1, params=[3,.1,10], options={'iterate': True})

modwind_dir_light, throwaway = pynumdiff.smooth_finite_difference._smooth_finite_difference.butterdiff(x=unwrap_angle(df.D.iloc[21000:27000]), dt=.1, params=[3,.5,10], options={'iterate': True})
modwind_dir_mod, throwaway = pynumdiff.smooth_finite_difference._smooth_finite_difference.butterdiff(x=unwrap_angle(df.D.iloc[21000:27000]), dt=.1, params=[3,.1,10], options={'iterate': True})
```

```python
gridspec = dict(wspace=0.0, hspace=.2, width_ratios=[3, .5,.85,3,.5])
fig, ax = plt.subplots(nrows=3, ncols=5, figsize=(20,12), gridspec_kw=gridspec)
ax[0,2].set_visible(False)
ax[1,2].set_visible(False)
ax[2,2].set_visible(False)
t=np.linspace(0,600, 6000)

#zero speed, closed wind tunnel
ax[0,0].scatter(t, df.S2.iloc[4000:10000], s=10, color= 'k', linewidth=None, alpha=.8, 
                label=("standard deviation: %s" % np.round(np.std(df.S2.iloc[4000:10000]),3)))
ax[0,0].scatter(t, nowind_speed_light, s=10, color= 'blue', linewidth=None, alpha=.5)
ax[0,0].scatter(t, nowind_speed_mod, s=10, color= 'red', linewidth=None, alpha=.2)

#speed ~1.2m/s
ax[1,0].scatter(t, df.S2.iloc[12000:18000], s=10, color= 'k', linewidth=None, alpha=.8, 
                label=("standard deviation: %s" % np.round(np.std(df.S2.iloc[12000:18000]),3)))
ax[1,0].scatter(t, lowwind_speed_light, s=10, color= 'blue', linewidth=None, alpha=.5)
ax[1,0].scatter(t, lowwind_speed_mod, s=10, color= 'red', linewidth=None, alpha=.2)

#speed ~2.2 m/s
ax[2,0].scatter(t, df.S2.iloc[21000:27000], s=10, color= 'k', linewidth=None, alpha=.8, 
                label=("standard deviation: %s" % np.round(np.std(df.S2.iloc[21000:27000]),3)))

ax[2,0].scatter(t, modwind_speed_light, s=10, color= 'blue', linewidth=None, alpha=.5)
ax[2,0].scatter(t, modwind_speed_mod, s=10, color= 'red', linewidth=None, alpha=.2)




#direction at zero speed
ax[0,3].scatter(t, df.D.iloc[4000:10000]+145, s=10, color= 'k', linewidth=None, alpha=.8, #add arbitrary value to make easier to see on plot
                label=("standard deviation: %s" % np.round(np.rad2deg(astropy.stats.circstd(np.deg2rad(df.D.iloc[4000:10000]+145))),3)))
ax[0,3].scatter(t, rewrap_angle(nowind_dir_light, degrees=True)+145, s=10, color= 'blue', linewidth=None, alpha=.5)
ax[0,3].scatter(t, rewrap_angle(nowind_dir_mod, degrees=True)+145, s=10, color= 'red', linewidth=None, alpha=.2)

#direction at ~1.2m/s
ax[1,3].scatter(t, df.D.iloc[12000:18000]-180, s=10, color= 'k',  linewidth=None, alpha=.8, #add arbitrary value to make easier to see on plot
                label=("standard deviation: %s" % np.round(np.rad2deg(astropy.stats.circstd(np.deg2rad(df.D.iloc[12000:18000]-180))),3)))
ax[1,3].scatter(t, rewrap_angle(lowwind_dir_light,degrees=True)-180, s=10, color= 'blue', linewidth=None, alpha=.5)
ax[1,3].scatter(t, rewrap_angle(lowwind_dir_mod,degrees=True)-180, s=10, color= 'red', linewidth=None, alpha=.2)


#direction at ~2.2m/s
ax[2,3].scatter(t, df.D.iloc[21000:27000]-180, s=10, color= 'k',  alpha=.8, #add arbitrary value to make easier to see on plot
                label=("standard deviation: %s" % np.round(np.rad2deg(astropy.stats.circstd(np.deg2rad(df.D.iloc[21000:27000]-180))),3)))

ax[2,3].scatter(t, rewrap_angle(modwind_dir_light,degrees=True)-180, s=10, color= 'blue', linewidth=None, alpha=.5)
ax[2,3].scatter(t, rewrap_angle(modwind_dir_mod, degrees=True) -180, s=10, color= 'red', linewidth=None, alpha=.2)


#histogram distributions for speed
ax[0,1].hist(df.S2.iloc[4000:10000], bins=100, orientation = 'horizontal', color='grey', alpha=0.6)
ax[1,1].hist(df.S2.iloc[12000:18000],bins=10, orientation = 'horizontal', color='grey', alpha=0.6)
ax[2,1].hist(df.S2.iloc[21000:27000],bins=10, orientation = 'horizontal', color='grey', alpha=0.6)
ax[0,1].axis('off')
ax[1,1].axis('off')
ax[2,1].axis('off')


#histogram distributions for direction
ax[0,4].hist(df.D.iloc[4000:10000]+145, bins=10, orientation = 'horizontal', color='grey', alpha=0.6)
ax[1,4].hist(df.D.iloc[12000:18000]-180,bins=100, orientation = 'horizontal', color='grey', alpha=0.6)
ax[2,4].hist(df.D.iloc[21000:27000]-180, bins=100,orientation = 'horizontal', color='grey', alpha=0.6)
ax[0,4].axis('off')
ax[1,4].axis('off')
ax[2,4].axis('off')

ax[0,0].set_ylim(0,3)
ax[1,0].set_ylim(0,3)
ax[2,0].set_ylim(0,3)
ax[0,1].set_ylim(0,3)
ax[1,1].set_ylim(0,3)
ax[2,1].set_ylim(0,3)
ax[0,3].set_ylim(0,360)
ax[1,3].set_ylim(0,360)
ax[2,3].set_ylim(0,360)
ax[0,4].set_ylim(0,360)
ax[1,4].set_ylim(0,360)
ax[2,4].set_ylim(0,360)

ax[0,0].legend(loc='upper right')
ax[1,0].legend(loc='upper right')
ax[2,0].legend(loc='upper right')

ax[0,3].legend(loc='upper right')
ax[1,3].legend(loc='upper right')
ax[2,3].legend(loc='upper right')

                
                
ax[0,0].set_ylabel('Wind speed (m/s)')
ax[1,0].set_ylabel('Wind speed (m/s)')
ax[2,0].set_ylabel('Wind speed (m/s)')
ax[0,3].set_ylabel('Wind direction')
ax[1,3].set_ylabel('Wind direction')
ax[2,3].set_ylabel('Wind direction')

ax[0,0].set_xlabel('Time (s)')
ax[1,0].set_xlabel('Time (s)')
ax[2,0].set_xlabel('Time (s)')
ax[0,3].set_xlabel('Time (s)')
ax[1,3].set_xlabel('Time (s)')
ax[2,3].set_xlabel('Time (s)')

#plt.tight_layout(pad=0.2, w_pad=2, h_pad=.5)


fifi.mpl_functions.adjust_spines(ax[0,0], 
                                  ['left', 'bottom'], 
                                  yticks=[0,1,2,3],
                                  xticks=[0,100,200,300,400,500,600], 
                                  tick_length=5,
                                  spine_locations={'left': 3, 'bottom': 10}, 
                                  linewidth=1.5)
                                 
fifi.mpl_functions.adjust_spines(ax[1,0], 
                                  ['left', 'bottom'], 
                                  yticks=[0,1,2,3],
                                  xticks=[0,100,200,300,400,500,600], 
                                  tick_length=5,
                                  spine_locations={'left': 3, 'bottom': 10}, 
                                  linewidth=1.5)
                                 
fifi.mpl_functions.adjust_spines(ax[2,0], 
                                  ['left', 'bottom'], 
                                  yticks=[0,1,2,3],
                                  xticks=[0,100,200,300,400,500,600], 
                                  tick_length=5,
                                  spine_locations={'left': 3, 'bottom': 10}, 
                                  linewidth=1.5)      

fifi.mpl_functions.adjust_spines(ax[0,3], 
                                  ['left', 'bottom'],
                                  yticks=[0,90,180,270,360],
                                  xticks=[0,100,200,300,400,500,600], 
                                  tick_length=5,
                                  spine_locations={'left': 3, 'bottom': 10}, 
                                  linewidth=1.5)
                                 
fifi.mpl_functions.adjust_spines(ax[1,3], 
                                  ['left', 'bottom'], 
                                  yticks=[0,90,180,270,360],
                                  xticks=[0,100,200,300,400,500,600], 
                                  tick_length=5,
                                  spine_locations={'left': 3, 'bottom': 10}, 
                                  linewidth=1.5)
                                 
fifi.mpl_functions.adjust_spines(ax[2,3], 
                                  ['left', 'bottom'],
                                  yticks=[0,90,180,270,360],
                                  xticks=[0,100,200,300,400,500,600], 
                                  tick_length=5,
                                  spine_locations={'left': 3, 'bottom': 10}, 
                                  linewidth=1.5)      


evenyticks=np.arange(0, 360+10, 90)
yticklabels=[['$' + str(y) + '^{\circ}$' + '',] for y in evenyticks]
yticklabels=np.ravel(yticklabels)
ax[0,3].set_yticklabels(yticklabels)
ax[1,3].set_yticklabels(yticklabels)
ax[2,3].set_yticklabels(yticklabels)



# for the histograms 
ax[0,1].tick_params(axis='both',          # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    bottom=False,      # ticks along the bottom edge are off
    left=False,         # ticks along the top edge are off
    labelbottom=False, labelleft=False)

ax[1,1].tick_params(axis='both',          
    which='both',      
    bottom=False,      
    left=False,         
    labelbottom=False, labelleft=False)
ax[2,1].tick_params(axis='both',          
    which='both',      
    bottom=False,      
    left=False,         
    labelbottom=False, labelleft=False)
                                 
ax[0,4].tick_params(axis='both',          
    which='both',      
    bottom=False,      
    left=False,         
    labelbottom=False, labelleft=False)

ax[1,4].tick_params(axis='both',          
    which='both',      
    bottom=False,     
    left=False,        
    labelbottom=False, labelleft=False)
ax[2,4].tick_params(axis='both',         
    which='both',     
    bottom=False,      
    left=False,         
    labelbottom=False, labelleft=False)





#fig.subplots_adjust(wspace=0.2)
fifi.mpl_functions.set_fontsize(ax[0,0], 22)

fig.savefig('noisetest.svg', format='svg')
```
